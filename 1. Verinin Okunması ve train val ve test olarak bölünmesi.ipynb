## Herbir kategoriden 1000 adet örneklem seçilmesi

```python
import os
import shutil
import numpy as np

# Define the source and destination directories
data_dir = r'C:\Users\User\Desktop\Hacettepe 2023\Deep Learning\Proje\EuroSAT_RGB'
sample_dir = r'C:\Users\User\Desktop\Hacettepe 2023\Deep Learning\Proje\EuroSAT_RGB_sample'

# Define the categories
categories = ['AnnualCrop',
              'Forest', 
              'HerbaceousVegetation', 
              'Highway', 
              'Industrial', 
              'Pasture', 
              'PermanentCrop', 
              'Residential', 
              'River', 
              'SeaLake']

# Define the number of samples per category
samples_per_category = 1000

# Loop through the categories
for category in categories:
    # Define the source and destination directories for the current category
    src_dir = os.path.join(data_dir, category)
    dst_dir = os.path.join(sample_dir, category)
    os.makedirs(dst_dir, exist_ok=True)
    
    # Count the total number of files in the source directory
    total_files = len(os.listdir(src_dir))
    
    # Calculate the number of files to copy
    num_files_to_copy = min(total_files, samples_per_category)
    
    # Copy the files to the destination directory
    file_names = os.listdir(src_dir)
    selected_files = np.random.choice(file_names, num_files_to_copy, replace=False)
    for file_name in selected_files:
        src_path = os.path.join(src_dir, file_name)
        dst_path = os.path.join(dst_dir, file_name)
        shutil.copy(src_path, dst_path)
```

## Kategorilere göre etiket ve yolların belirlenmesi

```python
# Define the list of categories in your dataset
categories = ['AnnualCrop',
              'Forest', 
              'HerbaceousVegetation', 
              'Highway', 
              'Industrial', 
              'Pasture', 
              'PermanentCrop', 
              'Residential', 
              'River', 
              'SeaLake']

# Initialize empty lists for the image paths and labels
image_paths = []
labels = []

# Loop over the categories and collect the image paths and labels
for label, category in enumerate(categories):
    category_dir = os.path.join(data_dir, category)
    for image_filename in os.listdir(category_dir):
        if image_filename.endswith('.jpg'):
            image_path = os.path.join(category_dir, image_filename)
            image_paths.append(image_path)
            labels.append(label)
```

## Verinin train validation ve test olarak split edilmesi ve dosyaların oluşturulması

```python
# Split the image paths and labels into training, validation, and testing sets
train_image_paths, val_test_image_paths, train_labels, val_test_labels = train_test_split(image_paths, labels, test_size=0.2, random_state=42)
val_image_paths, test_image_paths, val_labels, test_labels = train_test_split(val_test_image_paths, val_test_labels, test_size=0.5, random_state=42)

# Create the train, validation, and test directories and subdirectories
train_dir = os.path.join(data_dir, 'train')
val_dir = os.path.join(data_dir, 'val')
test_dir = os.path.join(data_dir, 'test')
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)
for category in categories:
    os.makedirs(os.path.join(train_dir, category), exist_ok=True)
    os.makedirs(os.path.join(val_dir, category), exist_ok=True)
    os.makedirs(os.path.join(test_dir, category), exist_ok=True)
```

## Görüntülerin ilgili dosyalara taşınmaları

```python
# Copy the training images to the train directory
for image_path, label in zip(train_image_paths, train_labels):
    category = categories[label]
    image_filename = os.path.basename(image_path)
    target_path = os.path.join(train_dir, category, image_filename)
    if not os.path.exists(target_path):
        shutil.copy2(image_path, target_path)

# Copy the validation images to the val directory
for image_path, label in zip(val_image_paths, val_labels):
    category = categories[label]
    image_filename = os.path.basename(image_path)
    target_path = os.path.join(val_dir, category, image_filename)
    if not os.path.exists(target_path):
        shutil.copy2(image_path, target_path)

# Copy the testing images to the test directory
for image_path, label in zip(test_image_paths, test_labels):
    category = categories[label]
    image_filename = os.path.basename(image_path)
    target_path = os.path.join(test_dir, category, image_filename)
    if not os.path.exists(target_path):
        shutil.copy2(image_path, target_path)
```
